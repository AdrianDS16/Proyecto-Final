```python
dfBalanceCR[["Carbon Mineral"]] = dfBalanceCR[["Carbon Mineral"]].astype("int")
dfBalanceCR.info()
```

```python
cat = np.linspace(min(dfBalanceCR["Carbon Mineral"]), max(dfBalanceCR["Carbon Mineral"]), 4)
cat
```

```python
nombres = ["bajo", "medio", "alto"]
dfBalanceCR["Carbon Mineral-bins"] = pd.cut(dfBalanceCR["Carbon Mineral"], bins= cat, labels = nombres, include_lowest=True)
dfBalanceCR[["Carbon Mineral", "Carbon Mineral-bins"]].head(5)
```

```python
fig, ax = plt.subplots()
ax.scatter (dfBalanceCR["Año"], dfBalanceCR["Carbon Mineral"], alpha = 0.5)

ax.set_xlabel("Año", fontsize=15)
ax.set_ylabel("Carbon Mineral", fontsize=15)
ax.set_title("Carbon Mineral a traves del tiempo")

ax.grid(True)
fig.tight_layout()

plt.show()
```
